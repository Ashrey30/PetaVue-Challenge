```python
import pandas as pd
import json
from datetime import datetime
```

```python
flight_cols = [
    "FlightDate",
    "Quarter",
    "Year",
    "Month",
    "DayofMonth",
    "DepTime",
    "DepDel15",
    "CRSDepTime",
    "DepDelayMinutes",
    "Origin",
    "Dest",
    "ArrTime",
    "CRSArrTime",
    "ArrDel15",
    "ArrDelayMinutes",
]
```

```python
airports = ["ATL", "CLT", "DEN", "DFW", "EWR", "IAH", "JFK", "LAS", "LAX", "MCO", "MIA", "ORD", "PHX", "SEA", "SFO"]

def load_flight_data(year):
    flight_data = pd.DataFrame()
    for month in range(1, 13):
        data_path = f"flight_data/{year}"
        file_path = f"{data_path}/On_Time_On_Time_Performance_{year}_{month}.csv"
        try:
            df = pd.read_csv(file_path)
            df = df[(df["Origin"].isin(airports)) & (df["Dest"].isin(airports))]
            df = df[flight_cols]
            flight_data = pd.concat([flight_data, df], ignore_index=True)
        except FileNotFoundError:
            print(f"File not found: {file_path}")
    return flight_data
```

```python
flight_data_2016 = load_flight_data(2016)
flight_data_2017 = load_flight_data(2017)

flight_data = pd.concat([flight_data_2016, flight_data_2017], ignore_index=True)
```

```python
weather_cols = [
    "date",
    "time",
    "airport",
    "tempF",
    "WindSpeedKmph",
    "WindDirDegree",
    "WeatherCode",
    "precipMM",
    "Visibility",
    "Pressure",
    "Cloudcover",
    "DewPointF",
    "WindGustKmph",
    "Humidity",
]
```

```python
def load_weather_data(year):
  weather_data = pd.DataFrame()
  for airport in airports:
    for month in range(1, 13):
      data_path = f"weather_data/{airport}"
      file_path = f"{data_path}/{year}-{month}.json"
      try:
        with open(file_path, "r") as f:
          try:
            data = json.load(f)
            weather_data_month = data.get('data')
            if weather_data_month:
              weather_list = weather_data_month.get('weather')
              if weather_list:
                df = pd.DataFrame(weather_list)
                df['airport'] = airport
                weather_data = pd.concat([weather_data, df], ignore_index=True)
          except json.JSONDecodeError:
            print(f"Error parsing JSON file: {file_path}")
            continue
      except FileNotFoundError:
        print(f"File not found: {file_path}")
  return weather_data
```

```python
weather_data_2016 = load_weather_data(2016)
weather_data_2017 = load_weather_data(2017)

weather_data = pd.concat([weather_data_2016, weather_data_2017], ignore_index=True)
```

```python
weather_data
```

```python
def merge_dicts(dicts):
    merged_dict = {}
    
    for dictionary in dicts:
        for key, value in dictionary.items():
            if key in merged_dict:
                merged_dict[key].append(value)
            else:
                merged_dict[key] = [value]
    
    for key, value in merged_dict.items():
        merged_dict[key] = value
    
    return merged_dict
```

```python
hourly_list = []

for i in range(len(weather_data)):
    hourly_list.append(merge_dicts(weather_data['hourly'][i]))
```

```python
df = pd.DataFrame(hourly_list)
```

```python
weather_data_joined = pd.concat([weather_data, df], axis=1)
```

```python
weather_data_joined.columns = weather_data_joined.columns.str.lower()
weather_cols_lower = [col.lower() for col in weather_cols]
```

```python
weather_data_joined = weather_data_joined[weather_cols_lower]
```

```python
weather_data_joined
```

```python
merged_data = pd.merge(flight_data, weather_data_joined, left_on=['FlightDate', 'Origin'], right_on=['date','airport'])
merged_data.to_csv('merged_data.csv', index=False)
```
